```python
# CONSTANTS and INPUT
MS = ... # Input in grams the mass of shots
MCS = ... # Input in grams the mass of calorimeter + stirrer
MCSW = ... # Input in grams the mass of calorimeter + stirrer + water
T_init_shots = ... # Input in deg C initial temp of shots; should be close to 100
T_init_water = ... # Input in deg C initial temp of water; should be close to Room Temperature
T_final = ... # Final temperature attained by the mixture
c_calorimeter = ... # Input the specific heat of calorimeter to find out the water equivalent in CGS ig idk man
LC_mass = 0.1 # grams
LC_temp = 0.1 # deg C

MW = MCSW - MCS # Mass of water in the calorimeter
MF = MCS * c_calorimeter # Water equivalent of calorimeter

# Specific heat of shots:
c_s =( (MW + MF) * (T_final - T_init_water) ) / ( MS * (T_init_shots - T_final))

# Error analysis:
delta_T_diff = np.sqrt(LC_temp**2 + LC_temp**2)
dt_numerator = T_final - T_init_water
dt_denominator = T_init_shots - T_final

delta_MW = np.sqrt(LC_mass**2 + LC_mass**2)

delta_MF = LC_mass * c_calorimeter

M_eff = MW + MF
delta_M_eff = np.sqrt(delta_MW**2 + delta_MF**2)

delta_MS = LC_mass

rel_M_eff = delta_M_eff / M_eff
rel_dt_num = delta_T_diff / dt_numerator
rel_MS = delta_MS / MS
rel_dt_den = delta_T_diff / dt_denominator

total_rel_error = np.sqrt(rel_M_eff**2 + rel_dt_num**2 + rel_MS**2 + rel_dt_den**2)

delta_c_s = c_s * total_rel_error
percent_error = total_rel_error * 100

print(f"Specific Heat (c_s): {c_s:.3f} cal/g/C") 
print(f"Uncertainty: {delta_c_s:.3f}")
print(f"Percentage Error: {percent_error:.2f}%")
```
