```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# INPUTS AND CONSTANTS
wavelength = np.array([...]) # Input the wavelengths of filter in nm
stopping_potential = np.array([...]) # Input the stopping potentials in V
LC_V = ... # Least Count of Voltmeter guy
LC_lambda = ... # Wavelength range of filter
e = 1.602e-19 # Charge of an electron in coulombs
c = 3e8 # Speed of light in m/s

# Calculations and other stuff:
wl_m = wavelength * 1e-9
delta_wl_m = LC_lambda * 1e-9

frequency = c/wl_m

delta_F = (c / (wl_m ** 2)) * delta_wl_m # Error in frequency

# Linregress stuff
fit = linregress(frequency, stopping_potential)

# Fit params
slope = fit.slope
intercept = fit.intercept
r_squared = fit.rvalue ** 2

# Error in slope and intercept
delta_slope = fit.stderr
delta_intercept = fit.intercept_stderr

# Derived quantities
h = slope * e
phi = -intercept # in eV

# Propagated errors:
delta_h = delta_slope * e
delta_phi = delta_intercept

print(f"Slope: {slope:.4e} +- {delta_slope:.4e} V.s")
print(f"Intercept: {intercept:.4f} +- {delta_intercept:.4f} V")
print(f"R^2: {r_squared:.4f}")
print(f"Planck's constant: {h:.4e} +- {delta_h:.4e} J.s")
print(f"Work Function (Phi): {phi:.4f} +- {delta_phi:.4f} eV")

plt.figure(figsize=(10,6))

plt.errorbar(frequency, stopping_potential, xerr=delta_F, yerr=LC_V, 
             fmt='bo', ecolor='blue', capsize=2, label='Data')

x_fit = np.linspace(frequency.min(), frequency.max(), 100)
y_fit = slope * x_fit + intercept

equation_label = f"Fit: $V = ({slope:.2e})\\nu {intercept:+.2f}$\n$R^2 = {r_squared:.4f}$"
plt.plot(x_fit, y_fit, 'k-', label=equation_label)

plt.xlabel("Frequency (Hz)")
plt.ylabel("Stopping Potential (V)")
plt.title("Stopping Potential vs Frequency")
plt.legend()
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()
```
