```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# INPUTS and CONSTANTS
m = ... #order of diffraction
d = ... # Grating spacing
left = np.array([...]) # Left angles
right = np.array([...]) # Right angles

LC_spectro = ... # Least count of spectrometer

n_f = 2 # Balmer series
n_i = np.array([3, 4, 5]) # Intial number

# Diffraction angle:
theta_deg = np.abs(left - right) / 2.0
theta_rad = np.radians(theta_deg)
delta_theta_Rad = np.radians(LC_spectro)

print("The diffraction angles are: ", theta_deg)

wl = (d * np.sin(theta_rad)) / m

delta_wl = (d * np.cos(theta_rad) / m) * delta_theta_Rad

# Plotting stuff
y_err = delta_wl / (wl**2)

x = 0.25 - (1 / (n_i**2))

y = 1 / wl

fit = linregress(x,y)

R_exp = fit.slope
delta_R = fit.stderr
r_Squared = fit.rvalue**2

print(f"Calculated wavelegnths (nm): {wl * 1e9}")
print(f"Value of Rydberg Constant: {R_exp:.5e} +- {delta_R:.5e} m^-1")
print("R_squared: ", r_Squared)

plt.figure(figsize=(10,6))

plt.errorbar(x, y, yerr=y_err, fmt='bo', ecolor='blue', capsize=4, label='Data')

x_fit = np.linspace(x.min() * 0.9, x.max() * 1.1, 100)
y_fit = R_exp * x_fit + fit.intercept

equation_label = f"Fit: $y = ({R_exp:.3e})x {fit.intercept:+.3e}$\n$R^2 = {r_Squared:.4f}$"

plt.plot(x_fit, y_fit, 'k-', label=equation_label)

plt.xlabel(r"$(\frac{1}{4} - \frac{1}{n_i^2})$")
plt.ylabel(r"$1/\lambda\ (\mathrm{m^{-1}})$")
plt.title("Determination of Rydberg Constant using Balmer Series")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
